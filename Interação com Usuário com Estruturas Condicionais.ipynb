```python

Nome = input("Qual o seu nome?")
Idade = input("Qual a sua idade?")
Peso = float(input("Qual o seu peso?"))
Altura = float(input("Qual a sua altura?"))
PesoIdeal = Altura*0.4
Ema = Peso - PesoIdeal
print("Seu nome é:",Nome)
print("Você tem:",Idade,"Anos")
print("Você Pesa:",Peso,"kg")
print("Você tem de altura",Altura,"M")
print("Logo seu Peso Ideal é de:",PesoIdeal,"Kg")
if Ema > 0:
  print('Você precisa emagrecer:', Ema,'Kg')
elif Ema < 0:
  roundedEma = round(Ema,2)
  print('Você precisa Ganhar:', neg(roundedEma),'Kg')
```
